# Reconoce la cantidad de Articulos y boton Ver Mas

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36")
driver = webdriver.Chrome(options=options)


# ‚úÖ Acceder al sitio
URL = "https://larepublica.pe/verificador"
driver.get(URL)
time.sleep(3)  # ‚úÖ Esperar carga inicial


botones = driver.find_elements(By.XPATH, '//*[@id="__next"]/div/div[2]/button')
print(f"Botones encontrados con XPath: {len(botones)}")
contador_clicks = 0
while True:
    try:
        boton_ver_mas = driver.find_element(By.XPATH, '//*[@id="__next"]/div/div[2]/button')
        driver.execute_script("arguments[0].click();", boton_ver_mas)  # ‚úÖ Forzar clic con JavaScript
        time.sleep(3)  # ‚è≥ Esperar que se carguen m√°s art√≠culos
        contador_clicks += 1  # ‚úÖ Contar cu√°ntas veces hicimos clic en "Ver m√°s"
    except:
        print("üö´ No hay m√°s art√≠culos disponibles o el bot√≥n no reaccion√≥.")
        break  # üöÄ Termina cuando ya no haya m√°s noticias


# ‚úÖ Buscar los art√≠culos despu√©s de cargar todos los disponibles
articulos = driver.find_elements(By.XPATH, '//article[contains(@class, "ListItem_content-article")]')
print(f"Art√≠culos encontrados despu√©s de hacer clic en 'Ver m√°s': {len(articulos)}")
print(f"Cantidad de botoes 'Ver m√°s' presionados encontrados: {(contador_clicks)}")

# ‚úÖ Extraer t√≠tulos y URLs
for articulo in articulos[:3]:  # ‚úÖ Solo mostramos los primeros 3 t√≠tulos
    titulo = articulo.find_element(By.XPATH, './/a/h3').text  # üîç Extraer t√≠tulo
    enlace = articulo.find_element(By.XPATH, './/a').get_attribute("href")  # üîó Extraer enlace
    print(f"üì∞ {titulo} ‚Üí {enlace}")

driver.quit()
```

# Reconoce las caracteristicas de la noticia

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36")
driver = webdriver.Chrome(options=options)

# ‚úÖ Acceder al sitio
URL = "https://larepublica.pe/verificador"
driver.get(URL)
time.sleep(3)  # ‚úÖ Esperar carga inicial

# ‚úÖ Obtener el primer art√≠culo
primer_articulo = driver.find_element(By.XPATH, '//article[contains(@class, "ListItem_content-article")]')
titulo = primer_articulo.find_element(By.XPATH, './/a/h3').text.strip()  # ‚úÖ Extraer t√≠tulo sin entrar en la noticia
url_articulo = primer_articulo.find_element(By.XPATH, './/a').get_attribute("href")  # ‚úÖ Obtener URL

# ‚úÖ Acceder a la p√°gina del art√≠culo
driver.get(url_articulo)
time.sleep(3)  # ‚úÖ Esperar que cargue

# ‚úÖ Extraer contenido, autor, fecha y veredicto desde dentro del art√≠culo
try:
    titulo = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "PictureInternal_picture__title__Dcne_"))
    ).text.strip()
except:
    titulo = "Sin t√≠tulo"

try:
    autor = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//a[contains(@href, "/autor/")]'))
    ).text.strip()
except:
    autor = "Desconocido"

try:
    fecha_tag = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "MainContentVerificador_main__date___qxWC"))
    )
    fecha = fecha_tag.text.strip()

    # üîπ Eliminar "Publicado:" si aparece en la fecha
    if "Publicado:" in fecha:
        fecha = fecha.replace("Publicado:", "").strip()

except:
    fecha = "Sin fecha"


try:
    contenido_tag = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "MainContentVerificador_content__notice__verificador__2LvaP"))
    )
    contenido = " ".join([p.text.strip() for p in contenido_tag.find_elements(By.TAG_NAME, "p")])

    # üîπ Cortar la parte innecesaria
    if "*Si deseas saber si una publicaci√≥n en las redes sociales" in contenido:
        contenido = contenido.split("*Si deseas saber si una publicaci√≥n en las redes sociales")[0].strip()

except:
    contenido = "Sin contenido"


try:
    veredicto = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//*[@id="__next"]/div/div[2]/p[8]/strong'))
    ).text.strip()
except:
    veredicto = "No encontrado"

# ‚úÖ Imprimir los datos obtenidos
print(f"üì∞ {titulo}\nüë§ Autor: {autor}\nüìÖ Fecha: {fecha}\n‚úÖ Veredicto: {veredicto}\nüìú Contenido: {contenido}\nüîó URL: {url_articulo}")

driver.quit()
```

# Scrapping de noticias

```python
# ‚úÖ Importar librer√≠as necesarias
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36")
driver = webdriver.Chrome(options=options)

# ‚úÖ Acceder al sitio
URL = "https://larepublica.pe/verificador"
driver.get(URL)
time.sleep(3)  # ‚úÖ Esperar carga inicial

# ‚úÖ Cargar m√°s art√≠culos haciendo clic en "Ver m√°s"
contador_clicks = 0
limite_clicks = 50  # Puedes ajustar este n√∫mero
while contador_clicks < limite_clicks:
    try:
        boton = WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/div/div[2]/button'))
        )
        driver.execute_script("arguments[0].click();", boton)
        time.sleep(2)
        contador_clicks += 1
    except:
        break

# ‚úÖ Obtener los enlaces de los art√≠culos
articulos = driver.find_elements(By.XPATH, '//article[contains(@class, "ListItem_content-article")]')
print(f"üîó Total de art√≠culos detectados visualmente: {len(articulos)}")

enlaces = []
for articulo in articulos:
    try:
        link = articulo.find_element(By.XPATH, './/a').get_attribute("href")
        if link and link.startswith("/"):
            link = "https://larepublica.pe" + link
        enlaces.append(link)
    except:
        pass

print(f"‚úÖ Total de enlaces extra√≠dos: {len(enlaces)}")

# ‚úÖ Lista donde se almacenar√°n los datos
datos_articulos = []

# ‚úÖ Extraer informaci√≥n desde cada art√≠culo
for url_articulo in enlaces:
    try:
        driver.get(url_articulo)
        time.sleep(3)

        # T√≠tulo
        try:
            titulo = WebDriverWait(driver, 5).until(
                EC.presence_of_element_located((By.CLASS_NAME, "PictureInternal_picture__title__Dcne_"))
            ).text.strip()
        except:
            titulo = "Sin t√≠tulo"

        # Autor
        try:
            autor = driver.find_element(By.XPATH, '//a[contains(@href, "/autor/")]').text.strip()
        except:
            autor = "Desconocido"

        # Fecha
        try:
            fecha = driver.find_element(By.CLASS_NAME, "MainContentVerificador_main__date___qxWC").text.strip()
            fecha = fecha.replace("Publicado:", "").strip()
        except:
            fecha = "Sin fecha"

        # Contenido
        try:
            contenido_tag = driver.find_element(By.CLASS_NAME, "MainContentVerificador_content__notice__verificador__2LvaP")
            parrafos = contenido_tag.find_elements(By.TAG_NAME, "p")
            contenido = " ".join([p.text.strip() for p in parrafos])
            if "*Si deseas saber si una publicaci√≥n en las redes sociales" in contenido:
                contenido = contenido.split("*Si deseas saber si una publicaci√≥n en las redes sociales")[0].strip()
        except:
            contenido = "Sin contenido"

        # Veredicto
        try:
            veredicto = driver.find_element(By.XPATH, '//*[@id="__next"]/div/div[2]/p[8]/strong').text.strip()
        except:
            veredicto = "No encontrado"

        # Guardar datos
        datos_articulos.append({
            "T√≠tulo": titulo,
            "Autor": autor,
            "Fecha": fecha,
            "Veredicto": veredicto,
            "Contenido": contenido,
            "URL": url_articulo
        })

    except Exception as e:
        print(f"‚ö†Ô∏è Error al procesar {url_articulo}:\n{e}")

# ‚úÖ Convertir a DataFrame y guardar
df = pd.DataFrame(datos_articulos)
SALIDA = "/content/drive/My Drive/Colab Notebooks/Scrapping_Peru.xlsx"
df.to_excel(SALIDA, index=False)
print(f"‚úÖ Archivo guardado en: {SALIDA} con {len(df)} art√≠culos")

# ‚úÖ Cerrar navegador
driver.quit()
```
